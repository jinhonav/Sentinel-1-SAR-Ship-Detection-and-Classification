```python
# ============================================================
# OpenSARShip2 Local Copy + Zip Save
# Run this on CPU runtime
# ============================================================

from google.colab import drive
from pathlib import Path
import shutil
import time
import os

drive.mount("/content/drive")

# ============================================================
# Paths
# ============================================================

SRC = Path(
    "/content/drive/MyDrive/SAR_AI_Ship_Detection/OpenSARShip2"
)

LOCAL_DIR = Path(
    "/content/OpenSARShip2"
)

ZIP_SAVE_PATH = Path(
    "/content/drive/MyDrive/SAR_AI_Ship_Detection/OpenSARShip2_local.zip"
)

# ============================================================
# Copy Drive -> /content
# ============================================================

if LOCAL_DIR.exists():
    shutil.rmtree(LOCAL_DIR)

print("Copying OpenSARShip2 from Drive to /content ...")
t0 = time.time()

shutil.copytree(SRC, LOCAL_DIR)

print(f"Copy finished: {(time.time() - t0) / 60:.1f} min")

# ============================================================
# Check copied files
# ============================================================

num_files = sum(1 for _ in LOCAL_DIR.rglob("*") if _.is_file())
print("Copied files:", num_files)

# ============================================================
# Zip local folder and save to Drive
# ============================================================

if ZIP_SAVE_PATH.exists():
    ZIP_SAVE_PATH.unlink()

print("Creating zip file in Drive ...")
t0 = time.time()

shutil.make_archive(
    str(ZIP_SAVE_PATH).replace(".zip", ""),
    "zip",
    LOCAL_DIR
)

print(f"Zip saved: {ZIP_SAVE_PATH}")
print(f"Zip time: {(time.time() - t0) / 60:.1f} min")
```
